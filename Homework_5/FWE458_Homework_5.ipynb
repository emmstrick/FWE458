```python
import matplotlib.pyplot as plt
import numpy as np

# Problem 1
x = np.linspace(0,10,100) #create data to plot
y = np.sin(x)

plt.plot(x,y, color="blue", linestyle="dashed") #add color and linestyle to plot
plt.title("Line Plot") #add title
plt.xlabel("X") #add x-axis title
plt.ylabel("Sin(X)") #add y-axis title
```

```python
import seaborn as sns

# Problem 2

tips = sns.load_dataset("tips") #load dataset
sns.histplot(data=tips, color="gray", x="total_bill", bins=20, stat="density") #plot histogram
sns.kdeplot(data=tips, x="total_bill", color="r", label="Kernal Density Estimate") #plot kde
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.legend()
```

```python
# Problem 3

fig, axes = plt.subplots(1,2, figsize=(10,5)) #create two plots

sns.boxplot(data=tips, x="time", y="tip", ax=axes[0]) #add boxplot to first subplot
axes[0].set_title("Tips vs Time of Day Box Plot")
axes[0].set_xlabel("Time of Day")
axes[0].set_ylabel("Tip Amount")
sns.violinplot(data=tips, x="time", y="tip", ax=axes[1]) #add violin plot to second subplot
axes[1].set_title("Tips vs Time of Day Violin Plot")
axes[1].set_xlabel("Time of Day")
axes[1].set_ylabel("Tip Amount")
```

```python
# Problem 4

sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker") #create scatterplot
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker") #regression line for each smoker status
```

```python
# Problem 5

tips_matrix = tips.drop(columns=["sex", "smoker", "day", "time"]) #drop non-numeric columns
correlation_matrix = tips_matrix.corr(method='pearson') # calculate correlation matrix
sns.heatmap(correlation_matrix, cmap="viridis", annot=True) # visualize correlation matrix on heatmap
```

```python
# Problem 6

sns.pairplot(data=tips, hue="time", diag_kind="kde") #create pair plot with time colored and diagonal kde
```
